```python
skip_columns = [
    'koi_disposition',
    'koi_fpflag_nt',
    'koi_fpflag_ss',
    'koi_fpflag_co',
    'koi_fpflag_ec'
]

to_drop = [
    'log_koi_period_err2', 
    'log_koi_time0bk_err2', 
    'log_koi_impact_err2',
    'log_koi_duration_err2',
    'log_koi_depth_err2',
    'log_koi_prad_err2',
    'log_koi_insol_err2',
    'log_koi_steff_err2',
    'log_koi_slogg_err2'#,
#     'log_koi_srad_err2',
#     'sq_koi_period_err2', 
#     'sq_koi_time0bk_err2', 
#     'sq_koi_impact_err2',
#     'sq_koi_duration_err2',
#     'sq_koi_depth_err2',
#     'sq_koi_prad_err2',
#     'sq_koi_insol_err2',
#     'sq_koi_steff_err2',
#     'sq_koi_slogg_err2',
#     'sq_koi_srad_err2' 
]

length = 0
trans_data = pd.DataFrame()

for (col, data) in exo2.items():
    
    log_label = 'log_' + str(col)
    sq_label = 'sq_' + str(col)
    
    if col in skip_columns:
        continue    
        
    elif data.min() == 0:   
        trans_data[log_label] = np.log(exo2[col] + 0.001)
        #trans_data[sq_label] = np.square(trans_data[log_label])
        
    elif data.max() == 0:
        trans_data[log_label] = np.log(np.absolute(exo2[col] + 0.001))
        #trans_data[sq_label] = np.square(trans_data[log_label])
        
    else:
        trans_data[log_label] = np.log(exo2[col])
        #trans_data[sq_label] = np.square(trans_data[log_label])
                        
trans_data = trans_data.drop(columns=to_drop)
```

```python
trans_data.describe()
```

```python
fig = plt.figure(figsize=(15, 12))

for i in range(1, len(trans_data.columns)):
    plt.subplot(9, 9, i)

    vals = np.size(trans_data.iloc[:, i].unique())
    
    if vals < 10:
        bins = vals
    else:
        vals = 10

    plt.hist(trans_data.iloc[:, i], bins=20, color='#3F5D7D')
    plt.xlabel(trans_data.columns[i])

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(15,60))

for i in range(1, len(trans_data.columns)):
    
    plt.subplot(20, 5, i)

    ax = sns.boxplot(exo2['koi_disposition'], trans_data.iloc[:, i])
    plt.xticks(rotation = 45)     

plt.tight_layout()
```

```python
exo2 = pd.concat([exo2, trans_data], axis=1)
exo2.to_csv('transformed_features.csv')
```

```python
corr = exo2.corr()

f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, cmap=cmap, vmax=.3, center=0, square=True,
           linewidths=.5, cbar_kws={'shrink': .5})
```
